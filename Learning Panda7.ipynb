Cleaning Data

```python
import pandas as pd #reading our data into pandas

df = pd.read_excel(r"C:\Users\ASUS\OneDrive\Desktop\LEARNING PYTHON, PANDAS\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df                      #get rid of duplicates in the data frame
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df                     #drop uncessessary columns 
```

```python
#df['Last_Name'] = df['Last_Name'].str.lstrip("...") #used to remove '...' on the leftside in the last name column
#df['Last_Name'] = df['Last_Name'].str.lstrip("/") #used to remove '/' on the leftside in the last name column
#df['Last_Name'] = df['Last_Name'].str.rstrip("_") #used to remove '_' on the rightside in the last name column
df['Last_Name'] = df['Last_Name'].str.strip('123./_') #used to remove ('.''/''_') at once in the last name column
df
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace('[^a-zA-Z0-9]','', regex = True) 
df #making phone number column a string then replacing the non-numeric values in phone number column with nothing
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df #formarting the phone number column using lambda
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','', regex = True)
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','', regex = True)
df #replace the Na, nan values with nothing
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand = True)
df #splitting the address column into street address, state, zipcode)
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Y','Yes')
df['Paying Customer'] = df['Paying Customer'].replace('N','No')
df #replace the Y with Yes in paying_customer column
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Y','Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N','No')
df #replace the Y with Yes in do_not_contact column
```

```python
df = df.replace('N/a','')
df = df.fillna('')
df # replace N/a with space and fill NaN with space
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] =='Yes':
        df.drop(x, inplace=True)

df #indexing rows where do not contact in yes and dropping the rows
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] =='':
        df.drop(x, inplace=True)

df#indexing rows where there is no phone number and dropping the rows

#df.dropna(subset = 'Phone_Number', inplace = True) #used to drop null values
```
