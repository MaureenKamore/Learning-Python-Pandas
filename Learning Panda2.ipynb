```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\Learning Python,file sorter,webscraping\world_population.csv")
```

```python
df
```

```python
df[df['Rank'] <=10] #shows data where rank is less or equal to 10
```

```python
df[df['Rank'] <=10].sort_values(by = 'Rank') #sorts out data 1 to 10 in order ie ascending order
```

```python
df[df['Rank'] <=10].sort_values(by = 'Rank', ascending =False) #sorts out data 10 to 1 in order ie descending order
```

```python
df[df['Rank'] <=10].sort_values(by = ['Country','Rank']) 
#sorts out data country being the first to be considered then the rank
```

```python
df[df['Rank'] <=10].sort_values(by = ['Capital','Continent'], ascending = [True,False]) 
#sorts out data capital being the first to be considered then the continent. #ordering capital in ascending order, continent descending order 
```

```python
specific_country = ['Brazil','China']
df[df['Country'].isin (specific_country)] #shows specific countries we are looking for in the Country column
```

```python
df[df['Continent'].str.contains('Africa')] #shows data contained in the continent column where string is'Africa'
```
