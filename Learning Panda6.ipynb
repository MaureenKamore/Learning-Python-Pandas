Panda Visualization


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\Learning Python,file sorter,webscraping\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available) #different styles available in the matplotlibrary

plt.style.use('seaborn-v0_8-deep') #choose the style to use from the list below
```

```python
df.plot() #plots a line graph by default
```

```python
df.plot(kind = 'bar') #plots a bar graph
```

```python
df.plot(kind = 'bar', subplots = True) #separates each bar individually
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating') #adds title
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores') #adds title to x and y
```

```python
df.plot(kind = 'bar', stacked = True) #plots a vertically stacked bar graph
```

```python
df ['Flavor Rating'].plot.bar() #plots a flavor rating bar graph
```

```python
df.plot.barh(stacked = True) #plots a horizontally stacked bar graph
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating') #plots a scatter graph with x and y specified
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300) 
#'s' rep the size of the scatter points, we can me them big or small
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', c = 'red') 
#'c' rep the color of the scatter points, 
```

```python
df.plot.hist(bins = 15) #plots a histogram, bins are numerical ranges or interval
```

```python
df.boxplot()
```

```python
df.plot.area() #plots an area graph
```

```python
df.plot.area(figsize = (10,6)) #increase the size of the area plot
```

```python
df.plot.pie(y = 'Flavor Rating') #plots a pie chart where we specify column we are using
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (15,10)) #increase the size
```
