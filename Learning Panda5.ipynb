Merge,Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\Learning Python,file sorter,webscraping\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\Learning Python,file sorter,webscraping\LOTR 2.csv")
df2
```

```python
df1.merge(df2) #shows inner join as a default
```

```python
df1.merge(df2, how = 'inner') #does the inner merge
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID') 
#Merges on the fellowshipId column, hence separating the firstnames into 2
```

```python
df1.merge(df2, how = 'outer')  #does the outer merge
```

```python
df1.merge(df2, how = 'left')  #does the left merge
```

```python
df1.merge(df2, how = 'right')  #does the right merge
```

```python
df1.merge(df2, how = 'cross')  #does the cross merge ie compares df1 data against each df2 data.
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right') 
#shows ouput but looks messy, join works well with index.
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how = 'outer', lsuffix = '_Left', rsuffix = '_Right') 
df3 #does an outer join but we have to set index first
```

```python
pd.concat([df1,df2]) #concatenate takes the data in df1 as it is and puts it on top of df2 as it is
```

```python
pd.concat([df1,df2],join = 'inner') #combines similar columns in df1 and df2
```

```python
pd.concat([df1,df2],join = 'outer') #combines all the columns in df1 and df2
```

```python
pd.concat([df1,df2],join = 'outer', axis =1) #combines all the columns in df1 and df2 and places them side by side in axis 1
```
